```python
for col in df_in.columns.tolist()[1:] :
    print(col)
    print(df_in[col].unique())
```

```python
#Splitting the values in 2 categories, nuemarical and categorical
cols_numerical = ["age", "bp", "sg", "al", "su", "bgr", "bu", "sc", "sod", "pot", "hemo", "pcv", "wc", "rc"]
cols_categorical = ["rbc", "pc", "pcc", "ba", "htn", "dm", "cad", "appet", "pe", "ane", "class"]
```

```python
import re
import numpy as np

def clean_cols_numerical(value_num):
    value_num = str(value_num)
    if value_num is not None:
        value_num = value_num.rstrip()
        value_num = re.sub(r"\t","", value_num)
        value_num = re.sub(r"\?","", value_num)
        # value_num = re.sub(r"^nan$",np.nan, value_num)
        # value_num = re.sub(r"^None$",np.nan, value_num)
        # value_num = re.sub(r"^$",np.nan, value_num)
        print(value_num)
    else:
        value_num = np.nan
    if value_num == "":
        value_num = np.nan
    return value_num


df_clean = df_in.copy()
for col_num in cols_numerical :
    df_clean[col_num] = [clean_cols_numerical(x) for x in df_clean[col_num]]
for col_num in cols_categorical :
    df_clean[col_num] = [clean_cols_numerical(x) for x in df_clean[col_num]]
```

```python
for col in df_clean.columns.tolist()[1:] :
    print(col)
    print(df_clean[col].unique())
```

```python
cols_of_variables = [j for i in [cols_numerical, cols_categorical] for j in i]
cols_of_variables
```

```python
from sklearn.model_selection import cross_val_score , train_test_split
from sklearn.metrics import classification_report
```

```python
X = df_clean.drop(["class","Index_AD"], axis=1)
y = df_clean['class']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
RANDOM_STATE = 0
for col in cols_numerical:
    X = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=RANDOM_STATE)
```

### Standardizing and formatting data

```python



for col in cols_numerical:
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
```

```python
from sklearn.preprocessing import LabelEncoder

lab = LabelEncoder()

for col in cols_categorical:
    X_train[col]= lab.fit_transform(X_train[col])
    X_test[col]= lab.fit_transform(X_test[col])
```

```python
import xgboost as xgb
from sklearn.metrics import classification_report, accuracy_score

# Step 2: Initialize XGBoost classifier
model = xgb.XGBClassifier()

# Step 3: Fit the model on training data
model.fit(X_train, y_train)

# Step 4: Make predictions on test data
y_pred = model.predict(X_test)

# Step 5: Compute and display classification metrics
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```
